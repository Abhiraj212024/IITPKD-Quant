```python
import sys
import os
# Adds the parent directory to the system path to allow imports from the lib folder
sys.path.append(os.path.abspath(os.path.join('..')))

from lib.DataCollector import DataCollector
from datetime import datetime, timedelta
```

```python
YEARS = 5
end_date = datetime.now()
start_date = end_date - timedelta(days = 365 * YEARS)
```

```python
stock_tickers = ['AAPL', 'MSFT', 'GOOGL']
data_collector = DataCollector(stock_tickers, start_date, end_date)
stock_data = data_collector.fetch_data()
```

```python
# Add technical indicators to each stock's DataFrame

stock_data = {ticker: data_collector.add_technical_indicators(df) for ticker, df in stock_data.items()}
```

```python
for ticker, df in stock_data.items():
    print(f"Plotting price data for {ticker}")
    data_collector.plot_price_data(ticker)

    print(f"Plotting OHLCV bars for {ticker}")
    data_collector.plot_ohlcv_bars(ticker)
```
